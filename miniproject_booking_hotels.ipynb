## Посмотрим на топ-5 стран по числу успешных бронирований 

```python
top5_successful = booking.query('is_canceled == 0')\
  .groupby('country', as_index = False).agg({'is_canceled': 'count'})\
  .sort_values('is_canceled', ascending = False).head(5)

top5_successful = top5_successful.rename(columns = {'is_canceled': 'num_of_successful'})

top5_successful
```

## На сколько ночей  в среднем бронируют отели разных типов?

```python
avg_nights = booking.groupby('hotel', as_index = False) \
  .agg({'stays_total_nights': 'mean'}).round(2)   

avg_nights = avg_nights.rename(columns = {'stays_total_nights': 'avg_num_of_nights'})

avg_nights
```

```python
t1 = booking.query('is_canceled == 1 and hotel == "City Hotel"') \
.groupby(['arrival_date_year', 'arrival_date_month'], as_index = False) \
.agg({'is_canceled':'count'}) \
.sort_values(['arrival_date_year', 'is_canceled'], ascending = [True, False])  \
.groupby('arrival_date_year').head(1)


t1
```

## Сколько раз приходилось менять забронированный номер из-за овербукинга?

```python
num_of_room_changes = booking.query('reserved_room_type != assigned_room_type') \
  .agg({'reserved_room_type': 'count'})

num_of_room_changes = num_of_room_changes.rename({'reserved_room_type': 'num_of_room_changes'})

num_of_room_changes
```

## В каком месяце чаще всего успешно оформляли бронь в 2016? Изменился ли он в 2017 году?

```python
most_successful_month_2016 = booking.query('(is_canceled == 0) and (arrival_date_year == 2016)') \
  .arrival_date_month.value_counts() \
  .sort_values(ascending = False).head(1)

most_successful_month_2016
```

```python
most_successful_month_2017 = booking.query('(is_canceled == 0) and (arrival_date_year == 2017)') \
  .arrival_date_month.value_counts()\
  .sort_values(ascending = False).head(1)

most_successful_month_2017
```

## Отели какого типа в среднем пользуются большей популярностью у клиентов с детьми?

```python
booking['total_kids'] = booking['children']+booking['babies']

most_kids_friendly = booking.groupby('hotel', as_index = False) \
  .total_kids.mean()\
  .sort_values('total_kids', ascending = False).round(2).head(1)

most_kids_friendly

```

##     Посмотрим как наличие детей связано с отменой бронирования

##     Для этого рассчитаем Churn Rate для группы клиентов с детьми и без детей и сравним коэффициенты

```python
booking['has_kids'] = booking['total_kids'] > 0
```

```python
booking['has_kids']
```

```python
churn_rate_with_kids = \
  round((booking.query('is_canceled == 1 and has_kids == True').shape[0] / booking.query('has_kids == True').shape[0])*100, 2)

f'Churn Rate для клинтов с детьми: {churn_rate_with_kids}%'
```

```python
churn_rate_without_kids = \
  round((booking.query('is_canceled == 1 and has_kids == False').shape[0] / booking.query('has_kids == False').shape[0])*100, 2)

f'Churn Rate для клинтов без детей: {churn_rate_without_kids}%'
```
